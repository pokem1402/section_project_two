```python
import pandas as pd
import numpy as np
import sys
import os
import pickle
import seaborn as sns
from tqdm.notebook import tqdm
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams['font.family'] = 'Malgun Gothic'

df = pd.read_csv("./dataset/new_bgg_data_after_eda2_new.csv", encoding="utf-8")
df.drop(columns=["Rank", "Id", "Name"], inplace=True)
```

```python
df_category = pd.read_csv("./dataset/bgg_data_category.csv", encoding="utf-8")
df_domain = pd.read_csv("./dataset/bgg_data_domain.csv", encoding="utf-8")
df_mechanics = pd.read_csv("./dataset/bgg_data_mechanics.csv", encoding="utf-8")
```

```python
cols = list(df_category.iloc[:, 5:].sum(axis=0).index[:10])
data = df_category.groupby(["Year"]).sum().drop(columns=["Id", "Rating", "Complexity","Ncategory"]).loc[:, cols]

sns.lineplot(data[(data.index >= 2000) & (data.index <= 2020)])
```

```python
cols = list(df_domain.iloc[:, 5:].sum(axis=0).index[:10])
data = df_domain.groupby(["Year"]).sum().drop(
    columns=["Id", "Rating", "Complexity", "Ndomains"]).loc[:, cols]

sns.lineplot(data[(data.index >= 2000) & (data.index <= 2020)])
```

```python
cols = list(df_mechanics.iloc[:, 5:].sum(axis=0).index[:10])
data = df_mechanics.groupby(["Year"]).sum().drop(
    columns=["Id", "Rating", "Complexity", "Nmechanics"]).loc[:, cols]

sns.lineplot(data[(data.index >= 2000) & (data.index <= 2020)])
```
